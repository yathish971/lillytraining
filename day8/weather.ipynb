```python
# import libraries
from matplotlib import pyplot as plt
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
avg_temp_city_a = [15, 17, 20, 24, 28, 30, 32, 31, 29, 26, 20, 16]
avg_temp_city_b = [10, 12, 15, 19, 24, 28, 30, 29, 26, 22, 15, 12]
```

```python
# plotting
plt.plot(months,avg_temp_city_a,label="banglore",color="g")
plt.plot(months,avg_temp_city_b,label="Delhi",color="y")

# Title

plt.title("Monthly Average temprature")
# Label
plt.xlabel("Months")
plt.ylabel("Temp in c")
# Legend
plt.legend()
plt.grid(color='r')
plt.show()
# Change grid color

```

```python
# Subpplots 
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.plot(months,avg_temp_city_a,label="banglore",color="g")
plt.title("Monthly Average temprature Banglore")
# Label
plt.xlabel("Months")
plt.ylabel("Temp in c")
# Legend

plt.grid(color='r')
plt.subplot(1,2,2)
plt.plot(months,avg_temp_city_b,label="Delhi",color="y")
plt.title("Monthly Average temprature Delhi")
# Label
plt.xlabel("Months")
plt.ylabel("Temp in c")
# Legend

plt.grid(color='r')


# Export
plt.savefig("Weather.png")
plt.show()
```
