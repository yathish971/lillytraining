```python
# import libraries    
from matplotlib import pyplot as plt
```

```python
# Monthly sales data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
region_a_sales = [10000, 12000, 8000, 15000, 11000]
region_b_sales = [8000, 9000, 11000, 9500, 12000]
```

```python
# plotting
plt.plot(months,region_a_sales,label="Region A")
plt.plot(months,region_b_sales,label="Region B")


# Title
plt.title("Sales Report")

# Label

# plt.xlabel("month")
# plt.ylabel("units")
# Legend
plt.legend()
# Change grid color
plt.grid(color="b")
```

```python
# Subpplots 
plt.figure(figsize=(12,6))
plt.subplot(2,1,1)
plt.plot(months,region_a_sales,label="Region A")


# Title
plt.title("Sales Report Region A")

# Label

# plt.xlabel("Months")
# plt.ylabel("units")

plt.subplot(2,1,2)
plt.plot(months,region_b_sales,label="Region B")
# Title
plt.title("Sales Report Region B")

# Label
#

# Export
plt.savefig("sales.report.png")
```
